```python
# 2
import re
song_path = {'fp' : ''}
def parse_m3u(file_path):
    with open(file_path, 'r', encoding = 'utf-8', errors = 'ignore') as f:
        content = f.read()
        
        
        pattern = r'#EXTINF:\d+,(.*?)\n(.*)'
        matches = re.findall(pattern, content)
        
        #change to 1 to get the file path instead or [0:] for both
        metadata = [match[0] for match in matches]
        path = [match[1] for match in matches]
        song_path['fp'] = path
        return metadata
    
m3u_file_path = r"C:\Users\AY\Desktop\- Musicolet Analysis\2024\Relapse Demo\ZY 11b. Relapse Demo February 24.m3u"
metadata = parse_m3u(m3u_file_path)
#print(file_paths)
```

```python
# 7
# Function for getting day, month and year in the right format from text
def txt_to_int_month(month):
    if month == 'January':
        month = '01'
    elif month == 'February':
        month = '02'
    elif month == 'March':
        month = '03'
    elif month == 'April':
        month = '04'
    elif month == 'May':
        month = '05'
    elif month == 'June':
        month = '06'
    elif month == 'July':
        month = '07'
    elif month == 'August':
        month = '08'
    elif month == 'September':
        month = '09'
    elif month == 'October':
        month = '10'
    elif month == 'November':
        month = '11'
    elif month == 'December':
        month = '12'
    return month


# May 2024
# The year, month are sliced with sliced with the pattern above -
def repl_year_month_with_int(df):
    y = df.loc[0, 'date']
    yr = y[-2:]
    year = f"{yr[0]}02{yr[1]}"
    month_txt = y[:-3]
    month = txt_to_int_month(month_txt)
    day = '01'
    ydm = f'{year}-{month}-{day}'
    df.loc[0:, 'date'] = ydm
    df['date'] = pd.to_datetime(df['date'], format = "%Y-%m-%d" )
    

                    
repl_year_month_with_int(jan_24)
#jan_24['date'] = pd.to_datetime(jan_24['date'], format = "%Y-%m-%d")
```

```python
#17
# Creating a function that can add genres from a primary table to each row in all dfs in a dictionary based on their id
# The first for loop grabs a dictionary, it sets the index to the column `id`, it does that to the primary table too.
# This was done to enable the use of the loc indexing to find similar rows in different dfs.
# Now I could have indexed the 'id' directly using loc, but there were duplicate id's which made the result lesser, so I 
# used a list to circumvent that. I got the id's in a list `id_list`, and I used this to index the `genral_temp` one at a time
# and got the corresponding genre appended in `genre_list`.
# I then inserted that list into the df, droped, renamed and changed the index

def add_genre_to_df_in_dict(dict, genral):
    # This for loops 
    for a in dict:
        temp_df = dict[a].set_index('id', drop = False)
        genral_temp = genral.set_index('id', drop = False)
        
        id_list = []
        for b in temp_df.index:
            id_list.append(b)
    
        genre_list = []
        for c in id_list:
            if c in genral_temp.index:
                genral_temp1 = genral_temp[~genral_temp.index.duplicated(keep = 'last')] #
                y = genral_temp1.loc[c, 'genre']
                genre_list.append(y)
            else:
                genre_list.append(np.nan)
    
        temp_df.insert(3, 'genre_1', genre_list)
        temp_df.drop(columns = 'genre', inplace = True)
        temp_df.rename(columns = {'genre_1' : 'genre'}, inplace = True)
        temp_df.index = [x for x in range(0, (len(temp_df.index)))]
        temp_df.explode('genre')
        dict[a] = temp_df


#add_genre_to_df_in_dict(clean_df, genral)
```

```python
# 21 Creating a function that aggregates a column that contains all unique values, performs a statistical analysis on it,
#    and the aggregated df is returned with the statistical analysis result added for each unique value

def unique_agg(df, unique_col, val_col, new_agg_col_name):
    df_agg = pd.DataFrame(df.groupby(unique_col)[val_col].sum()).sort_values(by = unique_col) 
    agg_val = [x for x in df_agg[val_col]]
    df_agg_gen = df.drop_duplicates(subset = unique_col, keep = 'first').sort_values(by = unique_col)
    df_agg_gen.insert(4, new_agg_col_name, agg_val)
    df_agg_gen.sort_values(by = new_agg_col_name, ascending = False, inplace = True)
    df_agg_gen.drop(columns = val_col, inplace = True)
    df_agg_gen.index = [x for x in range(0, len(df_agg_gen.index))]
    return df_agg_gen

# df_agg = pd.DataFrame(full_prelim.groupby('id')['score'].sum()).sort_values(by = 'id') 
# agg_val = [x for x in df_agg['score']]
# df_agg_gen = full_prelim.drop_duplicates(subset = 'id', keep = 'first').sort_values(by = 'id')
# df_agg_gen.insert(4, 'total_score', agg_val)
# df_agg_gen.sort_values(by = 'total_score', ascending = False, inplace = True)
# df_agg_gen.drop(columns = 'score', inplace = True)
# df_agg_gen.index = [x for x in range(0, len(df_agg_gen.index))]

#unique_agg(month_24_f, 'date', 'score', 'total_score')
unique_agg(monthly_prelim, 'id', 'score', 'total_score').head(10)
```

```python
# 28 THE CONSECUTIVE ADVANCED GRADING SYSTEM FUNCTION - Songs get points based on consequtive appearances in playlists

def consec_grade(dict_with_df, score_index, base_score, df_to_update, consec_grade_col, max_consec_col, total_col):
    cons_key = [f'cons{x}' for x in range(0, (len(dict_with_df)))]
    cons_value = [0 for x in range(0, len(cons_key))]
    cons_pair = [*zip(cons_key,cons_value)]
    cons_denotators = {x:y for (x,y) in cons_pair}
    
    checking_months = {}
    for month in dict_with_df:
        if month == 'September 24':
            pass
        else:
            checking_months[month] = dict_with_df[month]
    total_songs = concat_df_in_dict(dict_with_df)
    unique_songs = [x for x in total_songs['id'].unique()]
    
    count = 0
    for song in unique_songs:
        count += 1
        truth_list = []
        cons_denotators = {x:y for (x,y) in cons_pair} # I needed to keep regenerating the dictionary cos the values kept compunding in the static dictionary I first created
        
        for month in checking_months:
            month_to_check = checking_months[month]
            month_to_check_ids = [x for x in month_to_check['id'].unique()]
            if song in month_to_check_ids:
                truth_list.append(True)
            else:
                truth_list.append(False)
                
        index = 0
        for bool in truth_list:
            key = cons_key[index]
            if bool == True:
                cons_denotators[key] += 1       
            else:
                index += 1
    
        mag_values_list = [x for x in cons_denotators.values()]
        total_mag = sum(mag_values_list) # There might be no points for this
        mag_cons_max = max(mag_values_list)
        
        while 1 in mag_values_list:
            mag_values_list.remove(1)
        
        #cons_mag_occurences = [2, 3, 4, 5, 6, 7, 8, 9, 10]
        
        cons_mag_occurences = [x for x in range(2, (len(dict_with_df) + 1))]
        scoring_system = []
        score = base_score
        count = 0
        while count != len(cons_mag_occurences):
            scoring_system.append(round(score, 2))
            count += 1
            score += score_index #/ score_factor
        # return scoring_system
        
        
        occurence_x_scores = [*zip(cons_mag_occurences, scoring_system)]
        mag_scores = {x:y for x,y in occurence_x_scores}
        
        cons_mag_score = 0
        for cons_mag_value in mag_values_list:
            for num_key in mag_scores:
                if num_key == cons_mag_value:
                    cons_mag_score += mag_scores[num_key]

        
        df_to_update.loc[song, consec_grade_col] = cons_mag_score
        df_to_update.loc[song, max_consec_col] = mag_cons_max
        df_to_update.loc[song, total_col] = total_mag
        


consec_grade(mag_24, 1.729, 1.729, advan_df, 'cons_mag_score', 'mag_cons_max', 'total_mag')
consec_grade(rel_24, 1.629, 1.629, advan_df, 'cons_rel_score', 'rel_cons_max', 'total_rel') #0.9
consec_grade(reld_24, 1.829, 1.829, advan_df, 'cons_reld_score', 'reld_cons_max', 'total_reld') #0.61
consec_grade(month_24, 1.129, 1.129, advan_df, 'cons_month_score', 'month_cons_max', 'total_month')

advan_df.sort_values(by = 'cons_mag_score', ascending = False).head(8)
```

```python
# 29 THE POSITIONAL ADVANCED GRADING SYSTEM FUNCTION (updated) - songs get points in ranges for their positons in playlist. 
#   The playlist is arranged in descending order of plays
def positional_grade(dict_with_df, newees_rush_months, factor, dynamic_grade, grade_range, level_points, score_col):
    for month in dict_with_df:
        if month in newees_rush_months:
            level_point = np.array([x for x in level_points])
            level_pointz = factor * level_point
        else:
            level_pointz = np.array([x for x in level_points])
            
        if dynamic_grade == True:
            dynamic_range = {'a' : 30, 'b' : 40}
            if len(dict_with_df[month]) < 160:
                dynamic_range['b'] = 30
            elif len(dict_with_df[month]) > 170:
                dynamic_range['a'] = 40
            grade_range = [1, 3, 5, 5, 10, 15, 20, 25, dynamic_range['a'], dynamic_range['b']]

        continuous_range = []
        index = -1
        cont_index = -2
        for x in grade_range:
            index += 1
            cont_index += 1
            if index == 0:
                continuous_range.append(x)
            else:
                x += continuous_range[cont_index]
                continuous_range.append(x)
                                                                                                                                                        
        
        ind = -1
        standard_range = []
        for x in continuous_range:
            ind += 1
            if ind == 0:
                standard_range.append((x-1, x))
            elif continuous_range.index(x) == (len(continuous_range) - 1):
                standard_range.append((continuous_range[ind - 1], x + 1))
            else:
                standard_range.append((continuous_range[ind - 1], x))
        

        #level_points = [9.85, 8.35, 7.15, 5.95, 4.75, 3.85, 2.95, 2.05, 1.45, 0.85]
        points_n_range = [*zip(standard_range, level_pointz)]
        points_range_dict = {(x,y) : z for (x,y),z in points_n_range}
    
    
        for (x,y) in standard_range:
            for ind_loc in range(x,y):
                score = points_range_dict[(x,y)]
                if ind_loc >= len(dict_with_df[month]):
                    break
                else:
                    dict_with_df[month].loc[ind_loc, score_col] = score

#grade_range = [1, 4, 3, 5, 6, 7, 8, 9, 23]
#level_points = [1, 2, 3, 4, 5, 6, 7, 8, 9]
level_points = np.array([9.85, 8.35, 7.15, 5.95, 4.75, 3.85, 2.95, 2.05, 1.45, 0.85])
positional_grade(month_24, ['January 24', 'May 24', 'June 24', 'October 24', 'November 24', 'December 24'], 1.529, True, [], level_points, 'pos_grade')
#positional_grade(month_24, [], 1.529, True, [], level_points, 'pos_grade')
month_24['June 24'].sort_values(by = 'pos_grade', ascending = False).head(8)
#def positional_grade(dict_with_df, newees_rush_months, factor, dynamic_grade, grade_range, level_points, score_col):
```
